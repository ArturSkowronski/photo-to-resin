# photo-to-resin — galeria przykładów

Ten notebook przeprowadza **pięć scen przez cały pipeline**: wygenerowanie zdjęcia →
mapa wysokości → model 3D (STL) → renderingi (symulacja podświetlenia, relief, widok 3D).
Wyniki lądują w `site/przyklady/` i zasilają galerię na stronie reklamowej.

Uruchomienie całości: `Run All` albo headless:
`jupyter nbconvert --to notebook --execute --inplace notebooks/przyklady.ipynb`

```python
from photo_to_resin.examples import build_all

manifest = build_all("../site/przyklady")
for wpis in manifest:
    print(
        f"{wpis['title']:<22} płytka {wpis['plate_mm'][0]} × {wpis['plate_mm'][1]} mm"
        f" · {wpis['triangles']:,} trójkątów · {wpis['stl_mb']} MB"
    )
```

## Podgląd: zdjęcie → litofan pod światło → widok 3D

```python
from pathlib import Path

from IPython.display import HTML, display

katalog = Path("../site/przyklady")
for wpis in manifest:
    slug = wpis["slug"]
    display(HTML(
        f"<h3>{wpis['title']}</h3>"
        + "".join(
            f'<img src="../site/przyklady/{slug}-{rodzaj}.png" width="260" style="margin:4px">'
            for rodzaj in ("zdjecie", "podswietlony", "3d")
        )
    ))
```
